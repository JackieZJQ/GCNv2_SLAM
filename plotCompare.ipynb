```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
results = pd.read_csv("compareResults.csv")
print(results)
```

```python
fr1_xyz_method1 = np.asarray(results.iloc[1:21,0:5].values)
fr1_xyz_method1 = fr1_xyz_method1.astype(np.float)
print(type(fr1_xyz_method1[0,0]))
print(fr1_xyz_method1.shape)
np.set_printoptions(suppress=True)
print(fr1_xyz_method1)
```

```python
fr1_xyz_method2 = np.asarray(results.iloc[1:21,7:12].values)
fr1_xyz_method2 = fr1_xyz_method2.astype(np.float)
print(type(fr1_xyz_method1[0,0]))
print(fr1_xyz_method2.shape)
np.set_printoptions(suppress=True)
print(fr1_xyz_method2)
```

```python
fr3_nnf_method1 = np.asarray(results.iloc[25:45,0:5].values)
fr3_nnf_method1 = fr3_nnf_method1.astype(np.float)
print(type(fr3_nnf_method1[0,0]))
print(fr3_nnf_method1.shape)
np.set_printoptions(suppress=True)
print(fr3_nnf_method1)
```

```python
fr3_nnf_method2 = np.asarray(results.iloc[25:45,7:12].values)
fr3_nnf_method2 = fr3_nnf_method2.astype(np.float)
print(type(fr3_nnf_method2[0,0]))
print(fr3_nnf_method2.shape)
np.set_printoptions(suppress=True)
print(fr3_nnf_method2)
```

```python
fr1_floor_method1 = np.asarray(results.iloc[49:69,0:5].values)
fr1_floor_method1 = fr1_floor_method1.astype(np.float)
print(type(fr1_floor_method1[0,0]))
print(fr1_floor_method1.shape)
np.set_printoptions(suppress=True)
print(fr1_floor_method1)
```

```python
fr1_floor_method2 = np.asarray(results.iloc[49:69,7:12].values)
fr1_floor_method2 = fr1_floor_method2.astype(np.float)
print(type(fr1_floor_method2[0,0]))
print(fr1_floor_method2.shape)
np.set_printoptions(suppress=True)
print(fr1_floor_method2)
```

```python
# fr1-desk
x = np.arange(1, 21, 1)
y1 = fr1_xyz_method1[0:20, 0]
y2 = fr1_xyz_method2[0:20, 0]

plt.figure(figsize=(10,6), dpi=80)

plt.plot(x, y1, color="blue", label="Method 1")
plt.plot(x, y2, color="red", label="Method 2")
plt.legend()
plt.grid()
plt.xlabel("Sequences")
plt.ylabel("Initlization Time(s)") 
plt.title("Initlization times ( 20 times on fr1-desk)") 
plt.show()

# fr3-nnf
x = np.arange(1, 21, 1)
y1 = fr3_nnf_method1[0:20, 0]
y2 = fr3_nnf_method2[0:20, 0]

plt.figure(figsize=(10,6), dpi=80)

plt.plot(x, y1, color="blue", label="Method 1")
plt.plot(x, y2, color="red", label="Method 2")
plt.legend()
plt.grid()
plt.xlabel("Sequences")
plt.ylabel("Initlization Time(s)") 
plt.title("Initlization times (20 times on fr3-nnf)") 
plt.show()

# fr1-floor
x = np.arange(1, 21, 1)
y1 = fr1_floor_method1[0:20, 0]
y2 = fr1_floor_method2[0:20, 0]

plt.figure(figsize=(10,6), dpi=80)

plt.plot(x, y1, color="blue", label="Method 1")
plt.plot(x, y2, color="red", label="Method 2")
plt.legend()
plt.grid()
plt.xlabel("Sequences")
plt.ylabel("Initlization Time(s)") 
plt.title("Initlization times (20 times on fr1-floor)") 
plt.show()
```
